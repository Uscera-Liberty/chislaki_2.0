Объявите тензор a размера (2, 3, 4) и тензор b размера (1, 8, 3) на GPU, иницилизируйте их случайно равномерно (.uniform_())
Затем измените форму тензора b, чтобы она совпадала с формой тензора a, получите тензор c
Переместите c на CPU, переместите a на CPU
Оберните их в Variable()
Объявите тензор L = torch.mean((c - a) ** 2) и посчитайте градиент L по c ( то есть  ∂𝐿∂𝑐)

Выведите градиент L по c

```python
from torch.autograd import Variable

print(torch.cuda.is_available())
device = torch.device("cude")

a = torch.empty((2, 3, 4), device=device).uniform_()
b = torch.empty((1,8,3), device=device).uniform_()
c = b.reshape(2, 3, 4)

a = a.cpu()
c = c.cpu()
a = Variable(a)
c = Variable(c, requires_grad=True)

L = torch.mean((c-a)**2)
L.backward()

print("Тензор a:")
print(a)

print("\nТензор c:")
print(c)

print("\nL:")
print(L)

print("\nГрадиент dL/dc:")
print(c.grad)
```

### Задача 5

Доработайте многослойную нейронную сеть из файла "4_3 mlp pytorch tasks".

```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt

# создаём данные для 3 классов
N = 100   # количество точек в каждом классе
D = 2     # количество признаков
K = 3     # количество классов

# массив для координат точек
X = np.zeros((N * K, D))

# массив для номеров классов
y = np.zeros(N * K, dtype=np.int64)

# генерируем 3 "спиральных" класса
for j in range(K):

    # индексы объектов текущего класса
    ix = range(N * j, N * (j + 1))

    # расстояние от центра
    r = np.linspace(0.0, 1, N)

    # угол + небольшой шум
    t = np.linspace(j * 4, (j + 1) * 4, N) + np.random.randn(N) * 0.2

    # записываем координаты
    X[ix] = np.c_[r * np.sin(t), r * np.cos(t)]

    # записываем номер класса
    y[ix] = j

# переводим данные в тензоры PyTorch
X_tensor = torch.FloatTensor(X)
y_tensor = torch.LongTensor(y)

# создаём нейросеть
model = nn.Sequential(
    nn.Linear(2, 128),
    nn.ReLU(),
    nn.Dropout(0.2),

    nn.Linear(128, 128),
    nn.ReLU(),
    nn.Dropout(0.2),

    nn.Linear(128, 64),
    nn.ReLU(),

    nn.Linear(64, 3)
)

# функция ошибки
loss_fn = nn.CrossEntropyLoss()

# оптимизатор
optimizer = optim.Adam(model.parameters(), lr=0.001)

# обучение
epochs = 2000

for epoch in range(epochs):

    optimizer.zero_grad()

    predictions = model(X_tensor)

    loss = loss_fn(predictions, y_tensor)

    loss.backward()

    optimizer.step()

    if epoch % 200 == 0:
        predicted_classes = torch.argmax(predictions, dim=1)
        accuracy = (predicted_classes == y_tensor).float().mean()

        print(
            f"Epoch {epoch:4d} | "
            f"Loss = {loss.item():.4f} | "
            f"Accuracy = {accuracy.item():.4f}"
        )

# итоговая точность
model.eval()

with torch.no_grad():
    final_predictions = model(X_tensor)
    final_classes = torch.argmax(final_predictions, dim=1)
    final_accuracy = (final_classes == y_tensor).float().mean()

print(f"\nFinal accuracy: {final_accuracy.item():.4f}")

# строим границы классификации
h = 0.02

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

xx, yy = np.meshgrid(
    np.arange(x_min, x_max, h),
    np.arange(y_min, y_max, h)
)

grid = np.c_[xx.ravel(), yy.ravel()]
grid_tensor = torch.FloatTensor(grid)

with torch.no_grad():
    grid_predictions = model(grid_tensor)
    Z = torch.argmax(grid_predictions, dim=1)

Z = Z.numpy().reshape(xx.shape)

plt.figure(figsize=(10, 8))

plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.rainbow)

plt.scatter(
    X[:, 0],
    X[:, 1],
    c=y,
    cmap=plt.cm.rainbow,
    edgecolors="black",
    s=40
)

plt.title("Decision boundaries of improved MLP")
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```
