# Задание 4

Нормализовать элементы заданного одномерного массива таким образом, чтобы значения лежали в диапазоне [0, 1].

```python
arr = np.array([1, 5, 10, 15, 20])

min_val = np.min(arr)
max_val = np.max(arr)

if max_val == min_val:
    normalized_arr = [0]*len(arr)
else:
    normalized_arr = (arr - min_val) / (max_val - min_val)

print(normalized_arr)
```

# Задание 9

Создайте два массива одинаковой формы: первый состоит из четных чисел от 2 до 10, второй — из квадратов этих же чисел. Соедините оба массива горизонтально (по столбцам).

```python

arr1 = np.arange(2, 11, 2)
print(arr1)

arr2 = arr1**2

arr1_2d = arr1.reshape(-1, 1)
arr2_2d = arr2.reshape(-1, 1)

combined_arr = np.hstack((arr1_2d, arr2_2d))
print("\nResult\n")
print(combined_arr)
```

# Задание 11

Для двух векторов:
С и b= {0.6; 1.2; -1.8, 1.05} найти меру косинусного сходства.


```python
b = np.array([0.6, 1.2, -1.8, 1.05])

cos_sim = np.dot(C,b) / (np.linalg.norm(b) * np.linalg.norm(C))
print(cos_sim)
```
