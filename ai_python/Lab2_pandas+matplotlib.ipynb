#### Пояснения:
 - Определенное значение seed нужно, чтобы ответы у всех выполняющих
это задание были одинаковые и их можно было проверить (так как генерируются

 - Следует внимательнее использовать [ ] для выбора данных по нескольким условиям:

### Task 3.
Используя датасет Titanic, провести Exploratory Data Analysis (EDA) (с помощью описательной статистики), выявить основные факторы, повлиявшие на выживаемость пассажиров, рассчитать долю спасённых мужчин и женщин.

````python
titanic = pd.read_csv('titanic.csv')
print(titanic.info())
print(titanic.head())

print("Общая описательная статистика")
titanic.describe()
titanic.describe(include="object")

survive = round(titanic["Survived"].mean()* 100,2)
print(f"Общая доля выживших: {survive}%")

gender_surv = titanic.groupby('Sex')['Survived'].mean()
gender_surv = round(gender_surv * 100, 2)
print(gender_surv)

male = gender_surv['male']
female = gender_surv['female']
print(f"Доля спасенных мужчин: {male}")
print(f"Доля спасенных женщин: {female}")

class_surv = titanic.groupby('Pclass')['Survived'].mean()
print(round(class_surv * 100, 2))

age_surv = titanic.groupby('Survived')['Age'].mean()
print(round(age_surv, 2))

fare = titanic.groupby('Survived')['Fare'].mean()
print(round(fare, 2))

embarked_survival = titanic.groupby('Embarked')['Survived'].mean()
print(round(embarked_survival * 100, 2))
````

#### Вопросы:
     1) Ночные поступления денег (01-05 часов) в более чем 85% случаев являются мужскими.
     2) Посмотрев на долю мужчин в поступлениях средств (Income), можно сделать вывод, что количество поступлений средств женщинам в целом больше, чем мужчинам.
     3) Самые низкие траты в 3 часа ночи осуществляются в более 70% случаев женщинами.
     4) Существуют особые часы в мелких тратах, когда женщины тратят намного больше мужчин (>80%)
     5) Посмотрев на долю мужчин в максимальных тратах средств (Very High), можно сделать вывод, что количество высоких трат в каждый возможный час мужчин больше, чем у женщин.

##### Пояснения:
(\*) Обратите внимание, что в категории Very High Должны оказаться максимальные по модулю отрицательные транзакции. <br>
(\**) Например, для строки "0 10:23:26" час будет равен 10, а для строки "6 07:08:31"- 07. Можно воспользоваться функциями str.split() или str.find() и
функцией .apply(lambda x: x[])

````python

tr_mcc_codes = pd.read_csv("data/tr_mcc_codes.csv",sep=';')

tr_types = pd.read_csv("data/tr_types.csv",sep=';')

customers_gender_train = pd.read_csv("data/gender_train.csv")

df = transactions.merge(tr_mcc_codes, on="mcc_code", how="inner")
df = df.merge(tr_types, on="tr_type",how="inner")
df = df.merge(customers_gender_train, on="customer_id", how="left")

del tr_mcc_codes, tr_types, customers_gender_train
print(df.shape)

df['tr_day'] = df['tr_datetime'].apply(lambda x: int(x.split())[0])
df['tr_hour'] = df['tr_datetime'].str.split(':').str[0]

rashody = df['amount'] < 0
df['amount_bucket'] = pd.qcut(df.loc[rashody, 'amount'].abs(), 5, labels=['Very Low', 'Low', 'Middle', 'High', 'Very High'])
df['amount_bucket'] = df['amount_bucket'].cat.add_categories('Income')
df.loc[df['amount'] >= 0, 'amount_bucket'] = 'Income'

pivot_table = pd.pivot_table(
    df,
    values='gender',
    index='tr_hour',
    columns='amount_bucket',
    aggfunc='mean'
)


def plot_pivot_table(pivot):
    plt.figure(figsize=(12, 6))
    plt.imshow(pivot, aspect='auto')
    plt.xticks(range(len(pivot.columns)), pivot.columns, rotation=45)
    plt.yticks(range(len(pivot.index)), pivot.index)
    plt.colorbar()
    plt.tight_layout()
    plt.show()

plot_pivot_table(pivot_table)

````
